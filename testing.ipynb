```python
from typing import List
import datetime
import os
```

```python
class DataBase:
    def __init__(self, dir_path='src/database'):
        self.dir_path = dir_path
        
    def create(self, user_id: int, text: str, title: str | None) -> None:
        users = os.listdir(self.dir_path) # получаем список всех папок с пользователями
        user_folder_path = self.dir_path + f'/{user_id}' # получаем точный путь к папке, которую хотим создать
        
        if not (str(user_id) in users): # если нет пользователя
            os.makedirs(user_folder_path)
        
        if not title:
            title = datetime.datetime.now().__str__()
            
        notes = os.listdir(user_folder_path) # получаем список всех файлов, которые лежат в папке пользователя
        if f'{title}.txt' in notes:
            title = title + ' ' + datetime.datetime.now().__str__()
        
        note_path = user_folder_path + f'/{title}.txt'
        with open(note_path, 'w') as file: # w - write
            file.write(text)

        print(f'Создана заметка: {title}')
    
    def read(self, user_id: int, title: str) -> str | None:
        # если у пользователя есть своя папка
        # открываем заметку есть такая заметка
        print(f'Выдана заметка: {title}')
        # return notes
    
    def update(self, user_id: int, text: str, title: str) -> None:
        # если у пользователя есть своя папка
        # если есть заметка с title
        # то мы записывает текст 
        print(f'Изменена заметка: {title}')
        
    def delete(self, user_id: int, title: str) -> None:
        users = os.listdir(self.dir_path) # получаем список всех папок с пользователями        
        if str(user_id) in users: # если есть пользователь
            user_folder_path = self.dir_path + f'/{user_id}' # получаем точный путь к папке, которую хотим создать
            notes = os.listdir(user_folder_path) # получаем список всех файлов, которые лежат в папке пользователя
            if f'{title}.txt' in notes:
                note_path = user_folder_path + f'/{title}.txt'
                os.remove(note_path)
                print(f'Удалена заметка: {title}')
                return

        print('Заметка не удалена')
        
    def get_notes(self, user_id: int) -> List[str]:
        # если есть папка
        # если есть заметки
        # выдаем список заметок
        print('Выведены заметки')    
```
